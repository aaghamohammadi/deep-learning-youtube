Let's discuss a neuron in its simplest form. Imagine we have a single neuron, represented as $h_1$. The output $y$ is calculated as follows: $$y = 1 \times \phi_0 + h_1 \times \phi_1$$

```python
from IPython.display import Image

# Display image
Image(filename='images/007_neural_network_full.png', width=500)
```

In this equation, $\phi_0$ is the bias and $\phi_1$ is the weight associated with the neuron. For this discussion, let's assume that we feed the input $x$ to the neuron, which means $h_1 = x$. Therefore, we can express the output of a simple neural network with a bias of $\phi_0$ and a weight of $\phi_1$ as $$\phi_0 + \phi_1 h_1$$

```python
from IPython.display import Image

# Display image
Image(filename='images/007_neural_network.png', width=500)
```

```python
import tensorflow as tf

# Data
x = tf.constant([0.03, 0.19, 0.34, 0.46, 0.78, 0.81, 1.08, 1.18, 1.39, 1.60, 1.65, 1.90])
y = tf.constant([0.67, 0.85, 1.05, 1.0, 1.40, 1.5, 1.3, 1.54, 1.55, 1.68, 1.73, 1.6 ])

# Model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=(1,), activation="linear")
])

# Compile
model.compile(optimizer=tf.keras.optimizers.SGD(), loss=tf.keras.losses.mse)

# Train
model.fit(x, y, epochs=1000)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Predict
x_values = np.linspace(0, 2, 100)
y_pred = model.predict(x_values)

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, label='Actual')
plt.plot(x_values, y_pred, color='red', label='Predicted')
plt.xlim([0, 2])
plt.ylim([0, 2])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Actual vs Predicted')
plt.legend()
plt.show()
```

```python
weights, bias = model.layers[0].get_weights()

phi0 = bias
phi1 = weights[0]

print(phi0, phi1)
```
