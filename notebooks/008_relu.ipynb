So far, we've discussed a neural network that consists of a single hidden layer with only one neuron.

$$y = \phi_0 + \phi_1 h_1$$

Previously, we assumed that $h_1$ was linear:

$$h_1 = \theta_{10} + \theta_{11} x$$

However, what if we redefine the function $h_1$ to be non-linear? There are numerous possibilities, but the most commonly used is the *Rectified Linear Unit* or ReLU.

The formula for the ReLU function is:

$$h\left[z\right] = \max(0, z)$$

This function outputs the input directly if it's positive; otherwise, it outputs zero. It's widely used because it helps neural networks learn faster and perform better.


```python
import matplotlib.pyplot as plt
import tensorflow as tf

def relu(z):
    return tf.maximum(0, z)

def plot_relu(x, z, ylabel):
    plt.plot(x, relu(z))
    plt.title('ReLU Activation Function')
    plt.xlabel('x')
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()

x = tf.linspace(-3, 3, 1000)
z = 2 * x + 3
plot_relu(x, z, ylabel='ReLU(2x + 3)')
```

Let's explore the effects of adding neurons to a hidden layer that uses the ReLU activation function.

The output is formed by combining the hidden units:

$$y = \phi_0 + \phi_1 h_1 + \phi_2 h_2 + \phi_3 h_3$$

where each $h_i$ is defined as follows:

$$h_i = \text{ReLU}\left[\theta_{i0} + \theta_{i1} x\right]$$

```python
import matplotlib.pyplot as plt
import tensorflow as tf

def relu(z):
    return tf.maximum(0, z)

def plot_relu(x, z, ylabel, subplot_position):
    plt.subplot(2, 3, subplot_position)
    plt.plot(x, relu(z))
    plt.xlabel('x')
    plt.ylabel(ylabel)
    plt.grid(True)

x = tf.linspace(-3, 3, 1000)

# Define three hidden layers
z1 = 2 * x + 3
z2 = -3 * x - 1
z3 = 0.5 * x + 1

# Define weights for each hidden layer
phi0 = 0.5
phi1 = -1.2
phi2 = -0.8
phi3 = 0.6

# Sum the outputs of the three hidden layers with their respective weights
z_sum = phi0 + phi1 * relu(z1) + phi2 * relu(z2) + phi3 * relu(z3)

plt.figure(figsize=(10, 7))

# Plot each hidden layer
plot_relu(x, z1, r'ReLU(2x + 3)', 1)
plot_relu(x, z2, r'ReLU(-3x - 1)', 2)
plot_relu(x, z3, r'ReLU(0.5x + 1)', 3)

# Plot the sum of the three hidden layers
plt.subplot(2, 1, 2)
plt.plot(x, z_sum)
plt.title(r'$0.5 - 1.2 h_1 - 0.8 h_2 + 0.6 h_3$')
plt.xlabel('x')
plt.ylabel('Output, y')
plt.grid(True)

plt.tight_layout()
plt.show()
```

Given sufficient neurons, a shallow network has the ability to approximate any continuous one-dimensional function. This principle is known as the **Universal Approximation Theorem**.
