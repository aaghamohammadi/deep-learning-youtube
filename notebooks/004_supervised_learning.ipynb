**Supervised learning** algorithms establish a relationship between **input** data and the corresponding **output** prediction.

These algorithms can be thought of as a set of equations that map inputs to outputs.

The specific equation used is determined by **training data**, which consists of examples of input/output pairs.

supervised_learning.png

Here are some applications of supervised learning:
* Determining whether a computer tomography image indicates cancer or not.
* Translating an English sentence into French accurately.
* Forecasting next month's stock price based on this month's financial data.
* Converting an audio sample into a written transcription of the spoken words.
* Identifying and tagging the objects present in an image simultaneously.
